## Cuadrante Completo

```python
import folium

latitud_superior_izquierda = -31.432580
longitud_superior_izquierda = -64.252421
latitud_inferior_derecha = -31.451888
longitud_inferior_derecha = -64.242889

latitud_centro = (latitud_superior_izquierda + latitud_inferior_derecha) / 2
longitud_centro = (longitud_superior_izquierda + longitud_inferior_derecha) / 2

m = folium.Map(location=[latitud_centro, longitud_centro], zoom_start=12)

rectangle = folium.Rectangle(
    bounds=[
        (latitud_superior_izquierda, longitud_superior_izquierda),
        (latitud_inferior_derecha, longitud_inferior_derecha),
    ],
    color="blue",
    fill=True,
    fill_color="#3186cc",
).add_to(m)

folium.Marker([latitud_centro, longitud_centro], popup="Centro del Rectángulo").add_to(
    m
)

m.save("mapas_cuadrantes/cuadrante_completo.html")
m
```
